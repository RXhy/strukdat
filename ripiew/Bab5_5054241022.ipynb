1. Implementasikan fungsi pencarian dalam BST.


```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.val = key

class BST:
    def __init__(self):
        self.root = None

    def insert(self, key):
        self.root = self._insert(self.root, key)

    def _insert(self, root, key):
        if root is None:
            return Node(key)
        if key < root.val:
            root.left = self._insert(root.left, key)
        else:
            root.right = self._insert(root.right, key)
        return root
    
bst = BST()
bst.insert(5)
bst.insert(3)
bst.insert(8)
bst.insert(1)
bst.insert(4)
bst.insert(6)
```

```python
def search(self, key):
    return self._search(self.root, key)

def _search(self, root, key):
    if root is None or root.val == key:
        return root
    if key < root.val:
        return self._search(root.left, key)
    else:
        return self._search(root.right, key)

BST.search = search
BST._search = _search

print(bst.search(3)) #true
print(bst.search(7)) #false
```

2. Tambahkan fungsi untuk menghitung tinggi pohon.


```python
def height(self):
    return self._height(self.root)

def _height(self, root):
    if root is None:
        return -1  
    return 1 + max(self._height(root.left), self._height(root.right))

BST.height = height
BST._height = _height

print(bst.height())
```

3. Modifikasi traversal agar hasilnya disimpan dalam list.


```python
def preorder(self):
   result = []
   self._preorder(self.root, result)
   return result

def _preorder(self,root,result):
    if root:
        result.append(root.val)
        self._preorder(root.left, result)
        self._preorder(root.right, result)
        
        
def inorder(self):
    result = []
    self._inorder(self.root, result)
    return result

def _inorder(self, root, result):
    if root:
        self._inorder(root.left, result)
        result.append(root.val)
        self._inorder(root.right, result)


def postorder(self):
    result = []
    self._postorder(self.root, result)
    return result

def _postorder(self, root, result):
    if root:
        self._postorder(root.left, result)
        self._postorder(root.right, result)
        result.append(root.val)
        

        
BST.preorder = preorder
BST._preorder = _preorder
BST.inorder = inorder
BST._inorder = _inorder
BST.postorder = postorder
BST._postorder = _postorder

print("preorder traversal:", bst.preorder())
print("Inorder traversal:", bst.inorder())
print("postorder traversal:", bst.postorder())  
```

4. Buat model pohon keputusan untuk dataset sederhana.

```python
from sklearn import tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

X, y = load_iris(return_X_y=True)
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)

plt.figure(figsize=(12, 8))
tree.plot_tree(clf, filled=True)
plt.show()
```
